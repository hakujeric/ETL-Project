```python
# Import World Bank module
from pandas_datareader import wb
import numpy as np
import pandas as pd
```

```python
# Search for the indicator id for indicator name containing 'Greenhouse'
wb.search('greenhouse').iloc[:,:2]
```

```python
# Download the 'Total greenhouse gas emissions' data for all countries from 1970 to 2018 as panda dataframe
dat = wb.download(indicator='EN.ATM.GHGT.KT.CE', country='all',start=1970, end=2018)
dat.head(10)
```

```python
# Transform data: 
    #1. Reset index 
    #2. Set data in 'EN.ATM.GHGT.KT.CE' column as float
    #3. Set display data in 'EN.ATM.GHGT.KT.CE' column in numeric format
    #4. Groupby year so the data format is consistant with other datasets we have
dat.reset_index(level=0, inplace=True)
dat['EN.ATM.GHGT.KT.CE'].astype(float)
#dat['EN.ATM.GHGT.KT.CE'].map('{:,.2f}'.format)
pd.options.display.float_format = '{:,.2f}'.format
dat.groupby(['year']).mean()
```
