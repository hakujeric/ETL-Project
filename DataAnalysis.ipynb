```python
# Import the txt file with sea level data
seaLevel_df = pd.read_fwf('Resources/GMSL_TPJAOS_4.2_199209_201812.txt', header=None)
# seaLevel_df = seaLevel_df.columns ['altimeter type 0=dual-frequency  999=single frequency',
#                                      'year','number of observations',
#                      'number of weighted observations','life_exp','gdp_per_cap']
seaLevel_df.head()
```

```python
#Delete unneccary columns
seaLevel_df = seaLevel_df.drop(seaLevel_df.columns[[0,1,3,4,8,9,10,11]], axis=1)
seaLevel_df.head()
```

```python
# seaLevel_df.columns['year','number of weighted observations','GMSL','standard deviation of GMSL']
#                      'number of weighted observations','life_exp','gdp_per_cap']

seaLevel_df = seaLevel_df.rename(columns  = {2: "Year", 5:"number of weighted observations", 
                                             6:"GMSL",7:"standard deviation of GMSL"}) 
seaLevel_df.head()

# new_data = data.rename(columns = {"Team": "Team Name", 
#                                   "College":"Education", 
#                                   "Salary": "Income"})
```

```python
# seaLevel_df = seaLevel_df['Year'].str.split('.', 1).str
# seaLevel_df.head()
seaLevel_df['Year'] = seaLevel_df['Year'].astype(str)
seaLevel_df.dtypes
```

```python
seaLevel_df['Year'] = seaLevel_df['Year'].apply(lambda x : x.split('.')[0])
seaLevel_df.head()
```

```python
seaLevelbyYear_df = seaLevel_df.groupby(['Year']).mean()
seaLevelbyYear_df.head()
```
