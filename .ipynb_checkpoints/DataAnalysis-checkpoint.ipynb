```python
# Import the txt file with sea level data
seaLevel_df = pd.read_fwf('Resources/GMSL_TPJAOS_4.2_199209_201812.txt', header=None)
# seaLevel_df = seaLevel_df.columns ['altimeter type 0=dual-frequency  999=single frequency',
#                                      'year','number of observations',
#                      'number of weighted observations','life_exp','gdp_per_cap']
seaLevel_df.head()
```

```python
#Delete unneccary columns
seaLevel_df = seaLevel_df.drop(seaLevel_df.columns[[0,1,3,4,8,9,10,11]], axis=1)
seaLevel_df.head()
```

```python
# seaLevel_df.columns['year','number of weighted observations','GMSL','standard deviation of GMSL']
#                      'number of weighted observations','life_exp','gdp_per_cap']

seaLevel_df = seaLevel_df.rename(columns  = {2: "Year", 5:"number of weighted observations", 
                                             6:"GMSL",7:"standard deviation of GMSL"}) 
seaLevel_df.head()

# new_data = data.rename(columns = {"Team": "Team Name", 
#                                   "College":"Education", 
#                                   "Salary": "Income"})
```

```python

seaLevel_df['Year'] = seaLevel_df['Year'].astype(str)
seaLevel_df.dtypes
```

```python
seaLevel_df['Year'] = seaLevel_df['Year'].apply(lambda x : x.split('.')[0])
seaLevel_df.head()
```

```python
seaLevelbyYear_df = seaLevel_df.groupby(['Year']).mean()
seaLevelbyYear_df.head()
```

```python
#Import Global Means Land-Ocean Data
temp_df = pd.read_csv("Resources/GLB.Ts+dSST.csv")
temp_df.head()
```

```python
temp_df.drop(['J-D', 'D-N', 'DJF', 'MAM', 'JJA','SON'], axis=1, inplace=True)
temp_df.head()
```

```python
temp_df.set_index('Year',inplace=True)
temp_df['Average Temp'] = temp_df.mean(axis=1)
temp_df.head()
```

```python
tempYear_df = pd.DataFrame(temp_df['Average Temp'])
tempYear_df.head()
```

```python
# Import Ice data
ice_df = pd.read_excel("Resources/Sea_Ice_Index_Monthly_Data_by_Year_G02135_v3.0.xlsx")
ice_df.head()
```

```python
ice_df.set_index('Year', inplace=True)
ice_df.head()
```

```python
iceYear_df = pd.DataFrame(ice_df['Annual'])
iceYear_df = iceYear_df.rename(columns  = {'Annual': "Annual Sea Ice"}) 
iceYear_df.head()
```

```python
# Web Search
wb.search('greenhouse').iloc[:,:2]
```

```python
dat = wb.download(indicator='EN.ATM.GHGT.KT.CE', country='all',start=1970, end=2018)
dat.head()
```

```python
dat.reset_index(level=0, inplace=True)
dat['EN.ATM.GHGT.KT.CE'].astype(float)
greenHouse_df = dat.groupby(['year']).mean()
greenHouse_df.head()
```

```python
#Merge all Dataframe
# mergerData_df = pd.merge(seaLevelbyYear_df, tempYear_df, seaLevelbyYear_df, greenHouse_df, on='Year')
df1 = pd.merge(seaLevelbyYear_df,greenHouse_df, right_index=True, left_index=True)
df1.head()
```

```python
df1.reset_index(inplace=True)
df1.head()
```

```python
df1 = df1.rename(columns = {'index':'Year'})
df1['Year'] = df1['Year'].astype(int)
df1.head()
```

```python
df2 = pd.merge(iceYear_df, tempYear_df, right_index=True, left_index=True)
df2.head()
```

```python
df2.reset_index(inplace=True)
df2.head()
```

```python
mergeData_df = pd.merge(df1, df2, how='inner', on='Year')
mergeData_df.head()
```
